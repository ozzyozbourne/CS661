```python
np.random.seed(101)

df = pd.DataFrame(randn(5, 4), index=['A', 'B', 'C', 'D', 'E'], columns=['I', 'II', 'III', 'IV'])

df
```

```python
# Temporarily drop a row 
df.drop("B", axis=0)
```

```python
# Temporarily drop a column
df.drop("II", axis=1)
```

```python
# Permanenlty drop a row called C
df.drop("C", axis=0, inplace=True)
df
```

```python
# permamenlty drop a col called III
df.drop("III", axis=1, inplace=True)
df
```

```python
df["III"] = randn(len(df)) 
df
```
