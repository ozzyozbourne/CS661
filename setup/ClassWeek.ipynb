```python
x = np.linspace(0,10,50)
y = x **2
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, color = 'Red', marker = '>', linestyle = '-')
plt.title('X vs Y Values Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')

plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2,2,2)
plt.plot(y, x, 'b')

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Object Oriented Interface")
```

```python
fig = plt.figure(figsize=(8,7))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Main Plot")
axes1.set_xlabel("X Label")
axes1.set_ylabel("Y Label")

axes2.plot(x,y)
axes2.set_title("Sub Plot")
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()
fig.savefig('my_plot.png',dpi=200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x**2, label ='X Square',color='purple', linewidth=2,alpha=0.5,linestyle='--', marker='D',
          markersize=10,markerfacecolor ='yellow', markeredgewidth=3,markeredgecolor='blue')
axes.plot(x,x**3, label='X Cube',color='blue', lw=2,alpha=0.5,ls='--', marker='o' )
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Object Oriented Interface")
axes.legend(loc = 1)
```
